```python
import numpy as np
import fiona
from shapely.geometry import Point, Polygon, MultiPoint, MultiPolygon, shape, LineString

res_pltf_map = fiona.open('Data/Res_Str/Tikal_res_pltfm_grp_poly.shp')
```

```python
res_platform = []
for i in range(len(res_pltf_map)):
    res_platform.append(list(list(Polygon(shape(res_pltf_map[i]['geometry'])).centroid.coords)[0]))
```

# K-Means
Use the number of neighborhoods (18) as k-value

## Show the error vs. k-value

```python
from sklearn.cluster import KMeans
distortions = []
for i in range(1, 30):
    km = KMeans(n_clusters=i, init="k-means++", n_init=10,
                max_iter=3000, tol=1e-7, random_state=None)
    # n_init:设置初始样本中心的个数; max_iter: maximum num of iteration;
    # tol:设置算法的容错范围SSE(簇内误平方差);init:random-use Kmeans, default-use k-means++

    km.fit(res_platform)
    # get Sum of Squared Errors
    distortions.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.figure(facecolor='.6')
plt.plot(range(1, 30), distortions, marker="o")
plt.xlabel("No. of Clusters")
plt.ylabel("Sum of Squared Error")
plt.show()
```

```python
# Generate visually distance colors


def get_spaced_colors(n):
    max_value = 16581375  # 255**3
    interval = int(max_value / n)
    colors = [hex(I)[2:].zfill(6) for I in range(0, max_value, interval)]

    return tuple([(int(i[:2], 16), int(i[2:4], 16), int(i[4:], 16)) for i in colors])


# Reorganize the RGB data to hex
colors = []
for i in range(num_clusters+1):
    pass
#     colors.append('#%02x%02x%02x' % get_spaced_colors(num_clusters + 1)[1:][i])
colors = get_spaced_colors(num_clusters + 1)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics

from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(res_platform)

# Compute CBSCAN
eps_value = 0.1
db = DBSCAN(eps=eps_value, min_samples=7).fit(X)
# eps: max distance between two neighnoring samples
# min_samples: The number of samples (or total weight) in a neighborhood for a point to be considered as a core point. This includes the point itself.
```

```python
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True 
```

```python
# Show some data details

labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(res_platform, labels))
```

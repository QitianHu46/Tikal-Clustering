```python
import numpy as np
import fiona
from shapely.geometry import Point, Polygon, MultiPoint, MultiPolygon, shape, LineString

res_pltf_map = fiona.open('Data/Res_Str/Tikal_res_pltfm_grp_poly.shp')
```

```python
res_platform = []
for i in range(len(res_pltf_map)):
    res_platform.append(list(list(Polygon(shape(res_pltf_map[i]['geometry'])).centroid.coords)[0]))
```

# K-Means
Use the number of neighborhoods (18) as k-value

## Show the error vs. k-value

```python
from sklearn.cluster import KMeans
distortions = []
for i in range(1, 30):
    km = KMeans(n_clusters=i, init="k-means++", n_init=10,
                max_iter=3000, tol=1e-7, random_state=None)
    # n_init:设置初始样本中心的个数; max_iter: maximum num of iteration;
    # tol:设置算法的容错范围SSE(簇内误平方差);init:random-use Kmeans, default-use k-means++

    km.fit(res_platform)
    # get Sum of Squared Errors
    distortions.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.figure(facecolor='.6')
plt.plot(range(1, 30), distortions, marker="o")
plt.xlabel("No. of Clusters")
plt.ylabel("Sum of Squared Error")
plt.show()
```

## Plot for k = 18

```python
num_clusters = 18

km = KMeans(n_clusters=18, random_state=None)
km_labels = km.fit_predict(res_platform)
plt.figure(figsize=(10, 10), facecolor='0.6')
# draw residential platforms
plt.scatter(*zip(*res_platform), c=km_labels, s=5)
# draw central points for each cluster
plt.scatter(*zip(*km.cluster_centers_),
            c=[i for i in range(num_clusters)], s=100)


# draw boundaries for each cluster
from scipy.spatial import ConvexHull
for i in range(num_clusters):
    one_cluster = []
    for j in range(len(res_platform)):
        if km_labels[j] == i:
            one_cluster.append(res_platform[j])
    assert len(one_cluster) != 0
    hull = ConvexHull(one_cluster)
    for simplex in hull.simplices:
        plt.plot(np.asarray(one_cluster)[simplex, 0], np.asarray(
            one_cluster)[simplex, 1], 'k-')
    print(hull.area)


plt.title("K-Means \n Clustering of Resiodential Platforms \n num of Clusters: 18")
plt.show()
```

# DBSCAN

```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(res_platform)

# Compute CBSCAN
eps_value = 18e-2
db = DBSCAN(eps=eps_value, min_samples=8).fit(X)
# eps: max distance between two neighnoring samples
# min_samples: The number of samples (or total weight) 
# in a neighborhood for a point to be considered as a 
# core point. This includes the point itself.
```

The most important parameters for this algorithm, eps_value and min_samples are adjusted in order to get approximately 18 clusters, in order for comparability.

```python
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

db_labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
num_clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)
n_noise_ = list(db_labels).count(-1)
print('Estimated number of clusters: %d' % num_clusters)
print('Estimated number of noise points: %d' % n_noise_)
# print("Silhouette Coefficient: %0.3f"
#       % metrics.silhouette_score(res_platform, db_labels))
```

```python
num_clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10), facecolor='.6')
# Black removed and is used for noise instead.
unique_db_labels = set(db_labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_db_labels))]

for k, col in zip(unique_db_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (db_labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor=None, markersize=5)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'x', markerfacecolor=tuple(col),
             markeredgecolor=None, markersize=1)

    
# draw boundaries for each cluster
from scipy.spatial import ConvexHull
for i in unique_db_labels:
    one_cluster = []
    for j in range(len(X)):
        if db_labels[j] == i:
            one_cluster.append(X[j])
    assert len(one_cluster) != 0
    hull = ConvexHull(one_cluster)
    for simplex in hull.simplices:
#         plt.plot(np.asarray(one_cluster)[simplex, 0], np.asarray(
#             one_cluster)[simplex, 1], 'k-')    
        plt.plot(np.asarray(one_cluster)[simplex, 0], np.asarray(one_cluster)[simplex, 1], 'k-')  

    print(hull.area)
# for i in range(756):
#     if db_labels[i] == 0:
#         plt.scatter(X[i][0], X[i][1], s = 100)
    
        
plt.title('DBSCAN \n num of clusters: %d \n points with cover are\
core point, while others are phriphery points' % num_clusters)
```
